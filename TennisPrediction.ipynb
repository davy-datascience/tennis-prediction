## Data Collection


I found ATP match data from 2000 to 2017 on kaggle: https://www.kaggle.com/gmadevs/atp-matches-dataset

On the Acknowledgement section I found the Git repo where the up-to-date data is available: https://github.com/JeffSackmann/tennis_atp

From there I extracted data from 1968 to 2020.

## Data Preparation

The dataset contains 49 variables. There is **no independent variable**. So I added the boolean independent variable *p1_wins*, which at first will be always equal to 1 (True). 

To prevent from having an independent variable that has always the same value, I renamed winner and loser related variables by switching "*winner*" and "*loser*" prefixes with "*p1_*" and "*p2_*".

And then, for 50% of the dataset I switched "*p1_...*" and "*p2_...*" variables.
<br><br>
Several variables are related to data that is not available before the match as "p1_ace" (number of player 1 aces during the match).

This data should be excluded before modeling. However before removing those variables I will use some of them to build new features.
